# 4. click & send keys

- click

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.implicitly_wait(2)

driver.get('https://crawlingstudy-dd3c9.web.app/01/')

a_tag = driver.find_element(By.CSS_SELECTOR, 'a')
a_tag.click()
time.sleep(0.5)

p_tag = driver.find_element(By.CSS_SELECTOR, 'p')
print(p_tag.text)

time.sleep(3)

driver.quit()
```

- send keys

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.implicitly_wait(2)

driver.get('https://www.naver.com/')

a_tag = driver.find_element(By.CSS_SELECTOR, '#account a.MyView-module__link_login___HpHMW').click()
time.sleep(1)

id_tag = driver.find_element(By.CSS_SELECTOR, '#id').send_keys('test')

pw_tag = driver.find_element(By.CSS_SELECTOR, '#pw').send_keys('test')
time.sleep(1)

login_tag = driver.find_element(By.CSS_SELECTOR, '.btn_login.nlog-click').click()
time.sleep(3)

driver.quit()
```

# 5. 스크롤 내리기 

- 첫 번째 페이지 길이 갖고오기

```python
! pip install selenium
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
driver.implicitly_wait(2)

driver.get('https://www.musinsa.com/categories/item/001?device=mw')

height = driver.execute_script('return document.body.scrollHeight')
print(height)

time.sleep(1.5)
driver.quit()
```

- 스크롤 내리기 

```python
driver = webdriver.Chrome()
driver.implicitly_wait(2)

driver.get('https://www.musinsa.com/categories/item/001?device=mw')

for i in range(3):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(1.5)

    new_height = driver.execute_script('return document.body.scrollHeight')
    print(new_height)


time.sleep(3)

driver.quit()
```

# 6. page source 갖고오기 

```python
driver = webdriver.Chrome()
driver.get('https://www.musinsa.com/categories/item/001?device=mw')

response = driver.page_source
print(response)
print(type(response))
driver.quit()
```

# 7. 네이버 로그인 실습

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response, 'html.parser')

p_tag = soup.select_one('p')
print(p_tag.text.strip())
```
